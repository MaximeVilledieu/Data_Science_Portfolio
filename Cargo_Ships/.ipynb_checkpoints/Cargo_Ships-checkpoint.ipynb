#### Loading the data

We will organize what we need to build and preprocess our data in a global dataframe. 

```python
df_train_csv = pd.read_csv(os.path.join(path_cargo_dataset,"train/train.csv"))
```

```python
label_dict = {1: 'Cargo', 2: 'Military', 3: 'Carrier', 4: 'Cruise', 5: 'Tankers'}

df_train_csv['label'] = df_train_csv['category'].map(label_dict)
df_train_csv['path_image'] = df_train_csv['image'].apply(lambda img: os.path.join(os.path.join(path_cargo_dataset,"train/images/"), img))

df_train_csv
```

```python
df_train_csv.groupby('label').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

#### Data processing

```python
class ShipDataset(Dataset):
    def __init__(self, data_frame, transform=None):
        self.data_frame = data_frame
        self.transform = transform

    def __len__(self):
        return len(self.data_frame)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        img_path = self.data_frame.iloc[idx]['path_image']
        image = Image.open(img_path).convert('RGB')
        label = self.data_frame.iloc[idx]['label']

        if self.transform:
            image = self.transform(image)

        return image, label

data_transforms = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # Normalize
])

ship_dataset = ShipDataset(data_frame=df_train_csv, transform=data_transforms)

train_loader = DataLoader(ship_dataset, batch_size=32, shuffle=True)
```

```python
def imshow_grid(imgs, labels, nrows=3, ncols=3):
    fig, axs = plt.subplots(nrows, ncols, figsize=(8, 8))
    axs = axs.flatten()
    for img, label, ax in zip(imgs, labels, axs):
        img = img / 2 + 0.5
        np_img = img.numpy()
        ax.imshow(np.transpose(np_img, (1, 2, 0)))
        ax.set_title(f'{label}')
        ax.axis('off')
    plt.tight_layout()
    plt.show()

images, labels = next(iter(train_loader))
imshow_grid(images[:9], labels[:9])
```
